1) Prepare a classification model using Naive Bayes 
for salary data 

Data Description:

age -- age of a person
workclass	-- A work class is a grouping of work 
education	-- Education of an individuals	
maritalstatus -- Marital status of an individulas	
occupation	 -- occupation of an individuals
relationship -- 	
race --  Race of an Individual
sex --  Gender of an Individual
capitalgain --  profit received from the sale of an investment	
capitalloss	-- A decrease in the value of a capital asset
hoursperweek -- number of hours work per week	
native -- Native of an individual
Salary -- salary of an individual

```python
import numpy as np
import pandas as pd
#import label encoding
from sklearn import preprocessing
#importing Gaussian Naive Bayes model
from sklearn.naive_bayes import GaussianNB
```

```python
train = pd.read_csv('SalaryData_Train.csv')

train.head(1)
```

# Dropping out unnecessary columns from the dataset

```python
train.drop(['maritalstatus','relationship','race','capitalgain','capitalloss','educationno'],axis = 1,inplace=True)
```

```python
train.head(1)
```

```python
#creating label Encoder
le = preprocessing.LabelEncoder()
```

```python
# assigning columns to variables
x1 = train.iloc[:,0]
x2 = train['workclass']
x3 = train['education']
x4 = train['occupation']
x5 = train['sex']
x6 = train['hoursperweek']
x7 = train['native']
y = train['Salary']
```

```python
#converting strings labels into numbers
x1_en = le.fit_transform(x1)
x2_en = le.fit_transform(x2)
x3_en = le.fit_transform(x3)
x4_en = le.fit_transform(x4)
x5_en = le.fit_transform(x5)
x6_en = le.fit_transform(x6)
x7_en = le.fit_transform(x7)
```

```python
x2_en
```

```python
#combing all the x variables into single list of tuples
features = list(zip(x1_en,x2_en,x3_en,x4_en,x5_en,x6_en,x7_en))
features
```

```python
#creating Gaussian Classifier
model = GaussianNB ()
```

```python
model.fit(features,y)
```

```python
#now do prediction with testing data
```
